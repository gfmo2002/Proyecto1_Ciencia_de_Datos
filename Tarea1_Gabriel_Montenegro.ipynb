###Ejercicio 1

Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%. 

(Diversas opciones de solución: libre)

1.   Se genera una lista con el precio de 3 productos (números aleatorios entre 10 y 100)


```python
precios = []
for i in range(0,3):
  precios.append(random.randint(10,100))
precios
```

2.   Se obtiene la fecha actual y se genera una lista con los últimos 5 días (incluyendo el día actual). 
3.   Se itera sobre los 5 días y en cada uno se generan listas con la fecha, el producto, las unidades vendidas (números enteros aleatorios entre 1 y 20).

```python
fechas = pd.date_range(date.today(), periods=5).strftime('%d/%m/%y')
fecha = []
producto = []
unidades = []
precio = []
for i in fechas:
  for j in range(0,3):
    fecha.append(i)
    producto.append('Producto'+str(j+1))
    unidades.append(random.randint(1, 20))
    precio.append(precios[j])
print(fecha)
print(producto)
print(unidades)
print(precio)
```

4.   Se genera el Dataframe ventasdf con las listas fecha, producto, unidades, precio.
5.   Se agregan las columnas Total (unidades * precio) y Total con descuento (Total * 0.95)










```python
ventasdf = pd.DataFrame(list(zip(fecha, producto, unidades, precio)),
               columns =['Fecha', 'Producto', 'Unidades', 'Precio'])
#ventasdf
total = ventasdf.Unidades*ventasdf.Precio
#total
ventasdf['Total'] = total
#total_descuento = total*0.95
#total_descuento
ventasdf['Total_con_descuento'] = total*0.95
#df['new_col'] = mylist
ventasdf
```

###Ejercicio 2

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.

El resultado final debe ser una serie con los estadísticos.

1.   Se obtienen los nombres y salarios mensuales de 5 empleados. El resulatdo se guarda en un diccionario llamado "ingresos".

```python
ingresos = {}
for i in range(0,5):
  nombre = input("Ingrese nombre del empleado: ")
  salario = float(input("Ingrese el salario mensual del empleado: "))
  if salario < 0:
    print("El salario mensual debe ser un número mayor a 0")
  ingresos[nombre] = salario
  
print(ingresos)
```

2.   Se utilizan las funciones max y min sobre el diccionario y se calcula el promedio mediante la fórmula: suma(salarios) / cantidad(salarios) y se crea una lista con los 3 resultados.
3.   Se convierte la lista generada a una serie.

```python
ingresos_list = []
max_ingresos = max(ingresos.values())
#print(max_ingresos)
ingresos_list.append(max_ingresos)
min_ingresos = min(ingresos.values())
#print(min_ingresos)
ingresos_list.append(min_ingresos)
avg_ingresos = sum(ingresos.values()) / len(ingresos)
#print(avg_ingresos)
ingresos_list.append(avg_ingresos)
ser = pd.Series(ingresos_list, index = ['Max', "Min", "Mean"] )
print(ser)
```

###Ejercicio 3

Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla:

Mes      Ventas   Gastos
Enero    30500    22000
Febrero  35600    23450
Marzo    28300    18100
Abril    33900    35700
Mayo     42500    32450

A continuación, generar nuevas columnas:
- Columna 1 = Gasto/Ventas*100.
- Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
- Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
- Obtener el total de las ventas de los últimos 3 meses
- Obtener el total de gastos de febrero a abril








1.   Se crean 3 listas con la información de los meses, ventas y gastos. Luego se genera un dataframe "df" con el contenido de las 3 listas.
2.   Se genera una nueva columna llamada "Columna 1" en el Dataframe "df" con el resultado de la fórmula: Gastos / Ventas * 100
3.   Se genera la columna "Columna 2" con el texto "Meta superada" si el valor de las ventas es mayor o igual a 30,000 y "Meta no superada" si es menor a 30,000.
4.   Se genera la columna "Columna 3" con el resultado del sobregasto (Gastos - Ventas) si Gastos es mayor que Ventas y "0" si Gastos es igual o menor que Ventas.








```python
mes = []
ventas = []
gastos = []
df = []

mes = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [30500, 35600, 28300, 33900, 42500]
gastos = [22000, 23450, 18100, 35700, 32450]

df = pd.DataFrame(list(zip(mes, ventas, gastos)), columns=['Mes', 'Ventas', 'Gastos'])

df["Columna 1"] = df.Gastos / df.Ventas * 100
for i in df.Ventas:
  if i >= 30000:
    df["Columna 2"] = 'Meta superada'
  else:
    df["Columna 2"] = 'Meta no superada'

df["Columna 3"] = np.where(df["Gastos"] > df["Ventas"], df["Gastos"]-df["Ventas"], 0)

df
```

5.   Se calcula el total de ventas de los últimos 3 meses (marzo, abril, mayo).








```python
print("El total de ventas de los últimos 3 meses es: ", sum(df["Ventas"].iloc[2:]))
```

6.   Se calcula el total de gastos entre febrero y abril.








```python
print("El total de gastos de febrero a abril es: ", sum(df["Gastos"].iloc[1:4]))
```
